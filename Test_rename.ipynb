```python
from pmlb.dataset_lists import classification_dataset_names, regression_dataset_names
```

```python
from pathlib import Path
from os import path
from shutil import copyfile
```

```python
# mv and rename dataset
dtype = "classification"
new_classification_dataset_names = []
for d in classification_dataset_names:
    if d.count("-"):
        new_name = d.replace("-", "_")
        print("{}-->{}".format(d, new_name))
    else:
        new_name = d
    new_classification_dataset_names.append(new_name)
    new_dir = "datasets/{}".format(new_name)
    if not path.isdir(new_dir):
        Path(new_dir).mkdir(parents=True, exist_ok=True)
    old_file_path_1 = path.join("datasets/", dtype, d,"{}.tsv.gz".format(d))
    new_file_path_1 = path.join("datasets/", new_name, "{}.tsv.gz".format(new_name))
    copyfile(old_file_path_1, new_file_path_1)
    old_file_path_1 = path.join("datasets/", dtype, d,"{}.tsv.gz".format(d))
    new_file_path_1 = path.join("datasets/", new_name, "{}.tsv.gz".format(new_name))
    copyfile(old_file_path_1, new_file_path_1)
    old_readme_1 = path.join("datasets/", dtype, d,"README.md")
    new_readme_1 = path.join("datasets/", new_name,"README.md")
    #input file
    fin = open(old_readme_1, "rt")
    #output file to write the result to
    fout = open(new_readme_1, "wt")
    #for each line in the input file
    for line in fin:
        #read replace the string and write to output file
        fout.write(line.replace('# {}'.format(d), '# {}'.format(new_name)))
    #close input and output files
    fin.close()
    fout.close()
    
    
```
